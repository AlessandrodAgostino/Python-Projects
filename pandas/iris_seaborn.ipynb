```python
import pandas as pd
import seaborn
import pylab as plt

```

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
iris = pd.read_csv(url)
```

```python
iris.head()
```

```python
iris.info()
```

```python
from matplotlib import style
print(sorted(style.available))
style.use('default')
```

```python
seaborn.lmplot('sepal_length',
               'sepal_width',
               data=iris,
               col ='species')
```

```python
fg = seaborn.FacetGrid(data=iris,
                       col='species',
                       hue='species',
                       height=4,
                       aspect=0.9)
fg.map(plt.scatter,
       'sepal_length',
       'sepal_width', 
       s=50)

fg.map(seaborn.regplot,
       'sepal_length',
       'sepal_width', 
       scatter=False)
```

```python
seaborn.jointplot('sepal_length',
               'sepal_width', 
              data=iris, kind="kde");
```

```python
seaborn.boxplot('species',
                'sepal_length',
                data=iris,
                hue='species')
```

```python
seaborn.pairplot(iris, hue="species")
```

```python
g = seaborn.PairGrid(iris,
                     height=2,
                     hue='species')
g.map_diag(plt.hist, alpha=0.5)
g.map_offdiag(plt.scatter, alpha=0.75, s=20);
```
