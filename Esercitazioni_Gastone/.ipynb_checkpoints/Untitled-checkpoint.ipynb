```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pylab as plt 
```

## Plain fit to a dataset

```python
def datagen(x,eps):
    return np.sin(2*np.pi*x)+\
            st.norm.rvs(size = len(x), loc=0,scale=eps)
```

```python
#Print the real function
x = np.linspace(-1,1,1000)
y = np.sin(2*np.pi*x)
plt.plot(x,y,'g',linewidth=2,label='true func')

#simulation of measure
X = np.linspace(-1,1,10)
eps=0.2
Y = datagen (X,eps)
plt.scatter(X,Y,s=140,label='mydata',facecolors='w',edgecolor='b')
```

```python
#Try to make a fit using Numpy
m=8
p=np.polyfit(X,Y,m)
#prints me out the coefficients' values
#P is just an array of values, it's not yet the polynomial function
print(p)
```

```python
#I put up the polynomia using the cofficients 
yexp = np.poly1d(p)
plt.plot(x,yexp(x),'r--',label='poly '+str(m))
plt.legend()
plt.show
```

```python
#Here I print all the items together:
plt.plot(x,y,'g',linewidth=2,label='true func')
plt.scatter(X,Y,s=140,label='mydata',facecolors='w',edgecolor='b')
plt.plot(x,yexp(x),'r--',label='poly '+str(m))
plt.legend()
plt.show
```

## Measuring the RMS

```python
#I create a train and a test dataset
n_point = 10
X1 = np.linspace(0, 1, n_point)
train = datagen(X1,eps)

X = np.linspace(0, 1, n_point)
test = datagen(X,eps)

x = np.linspace(0,1,500)
y = np.sin(2*np.pi*x)

m = 8
p = np.polyfit(X1,train,m)
yexp = np.poly1d(p)

plt.plot(x,y,'g',linewidth=2,label='true func')
plt.scatter(X,test,s=140,label='test',facecolors='w',edgecolor='b')
plt.scatter(X1,train,s=140,label='train',facecolors='w',edgecolor='g')
plt.plot(x,yexp(x),'r--',label='poly '+str(m))
plt.legend()
plt.show
```

```python
#Define the normalized error function:
def RMS(y,yexp):
    s = np.sum((yexp-y)**2)
    n=len(y)
    return np.sqrt(s)/n
```

```python
mmax = 25 #max degree I want to test
testRMS = np.empty(mmax) #Allocating error containers
trainRMS = np.empty(mmax)
pp = np.zeros((mmax, mmax))#Coefficient matrix by varying m
M = np.arange(mmax)#Span of the coefficient

for m in M:
    #Make a polynomial fit of different degree on the same dataset
    p = np.polyfit(X1, train, m) 
    
    #Saving the coefficients: every column a degree 
    #every row a coefficient
    for i in np.arange(m+1):
        pp[i,m] = p[i]
    
    #Put up the polynomial function
    fexp = np.poly1d(p)
    #Make expectations about train dataset
    trainexp = fexp(X1)
    #Evaluate error between fitted values and expected values
    trainRMS[m] = RMS(train,trainexp)
    #Make the same process on the test dataset
    testexp = fexp(X)
    testRMS[m] = RMS(test, testexp)
```

```python
#Plot all my results superimposing scatterplots and plots
plt.scatter(M, trainRMS, s=140, facecolor='w', edgecolor='b')
plt.scatter(M, testRMS, s=140, facecolor='w', edgecolor='r')

plt.plot(M, trainRMS, 'ob-', label='train')
plt.plot(M, testRMS, 'or-', label='test')

plt.ylabel('RMS', fontsize=20)
plt.xlabel('M', fontsize=20)

plt.axvline(x=3, linestyle='--', label='Overfitting')

plt.legend(loc='best',fontsize=16)
plt.show()
```

## Second manner to avoid overfitting: inserting a penalization

```python
def sq_err_ridge(x,y,f, lamb= 0,p=()):
    #https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html
    from scipy.optimize import minimize 
    #Minimize requires a function and the vector of its arguments
    #I implement the cost function for a generic function f:
    def cost(args):
        #Insertion of the Ridge penalization
        return np.sum((y-f(x,args))**2)*0.5 + lamb*np.sum(args**2)*0.5
    return minimize(cost,p)
    #returns an OptimizeResult object. Whose attribute x contains the solutions

def sq_err(x,y,f,p=()):
    from scipy.optimize import minimize
    def cost(args):#i parametri vanno passati tutti insieme come un unico vettore
        return np.sum((y-f(x,args))**2)*0.5
    return minimize(cost,p) #restituisce un oggetto con vari attributi tra cui i valori dei parametri ottimizzati
    
    
#Need to define my polynomial function this way in to use minimize
def my_poly(x, args=()): #args has to be a tuple
    f = np.poly1d(args)
    return f(x)

#Exactly the same function as before
def RMS(y,yexp):
    s = np.sum((yexp-y)**2)
    n=len(y)
    return np.sqrt(s)/n #normalizzata
```

```python
#Don't know what she has done here
x = np.arange(10)
args = tuple(np.array([1,1,1,1,1,1]))
y = my_poly(x,args)

res = sq_err(x,y,my_poly,args).x
print('standard:',res)

res = sq_err_ridge(x,y,my_poly,lamb=1,p=args).x
print('penalized:',res)
```

```python
eps=0.2
n_point = 20
X1 = np.linspace(0, 1, n_point)
train = datagen(X1,eps)
X = np.linspace(0, 1, n_point)
test = datagen(X,eps)

#penalization parameter:
lamb=1e-3
mmax = 25
testRMS = np.empty(mmax)
trainRMS = np.empty(mmax)
pp = np.zeros((mmax, mmax))
M = np.arange(mmax)

for m in M:
    p = sq_err_ridge(X1, train, my_poly, lamb=lamb, p=np.zeros(m+1)).x
    
    for i in np.arange(m+1):
        pp[i,m] = p[i]

    fexp = np.poly1d(p)
    trainexp=fexp(X1)
    trainRMS[m] = RMS(train,trainexp)
    testexp=fexp(X)
    testRMS[m] = RMS(test, testexp)

plt.scatter(M, trainRMS, s=140, facecolor='w', edgecolor='b')
plt.scatter(M, testRMS, s=140, facecolor='w', edgecolor='r')
plt.plot(M, trainRMS, 'ob-', label='train')
plt.plot(M, testRMS, 'or-', label='test')
plt.ylabel('RMS', fontsize=20)
plt.xlabel('M', fontsize=20)
plt.legend(loc='best',fontsize=16)
plt.show()
```
