```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample_points = np.linspace(0,2*np.pi, 2**4+1)
function_point = np.linspace(0,2*np.pi, 1000)
under_function = np.sin(function_point)
```

```python
plt.plot(function_point, under_function,'-')
plt.title('Function sample')
plt.xlabel('x')
plt.ylabel('$sen(x)$')
```

```python
def add_normal_noise(sample, mean=0, var=0.12):
    return sample + np.random.normal(mean, var, len(sample))
```

```python
noised_sample = add_normal_noise(under_function)
```

```python
plt.scatter(sample_points, under_function, label='function',facecolors='w',edgecolor='b')
plt.scatter(sample_points, noised_sample, label='noised',facecolors='w',edgecolor='g')
plt.title('Function sample')
plt.xlabel('x')
plt.ylabel('$sen(x)$')
```

```python
m=3
p=np.polyfit(sample_points, noised_sample, m)
ext_pol=np.poly1d(p)
plt.plot(sample_points, ext_pol(sample_points), 'r--', label = "Poly -"+str(m))
plt.legend()
```

All in one plot

```python
plt.scatter(sample_points, under_function, label='function',facecolors='w',edgecolor='b')
plt.scatter(sample_points, noised_sample, label='noised',facecolors='w',edgecolor='g')
plt.plot(sample_points, ext_pol(sample_points), 'r--', label = "Poly -"+str(m))
plt.legend()
plt.title('Function sample')
plt.xlabel('x')
plt.ylabel('$sen(x)$')
```
